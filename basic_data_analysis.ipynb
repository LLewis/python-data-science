```python
import pandas as pd
import numpy as np
import seaborn as sns

polls_url = 'https://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'
#reads csv file into a Panda DataFrame
polls = pd.read_csv(polls_url)
print(polls)
print('***********************************************************')

#LIST enclosed in []
#drop() column method - Python List - drops columns in the list
polls.drop(columns=['cycle','branch'], inplace=True)
print(polls)
print('***********************************************************')

#TUPLE  enclosed in {}
polls.plot.line(xlim={'2016-09', '2016-11'})  #displays x,y graph


#DICTIONARY enclosed in [:]
polls.rename(columns={'adjpoll_clinton': 'Clinton',
                      'adjpoll_trump': 'Trump'})

#TWO SLICES USED IN A loc[] accessor
#the number 10 means to access every 10th row from 0 thru 100
#access every column from state to grade
polls.loc[0:100:10,'state':'grade']

#LIST COMPREHENSION
#start value 1900 , stop value 1920 (does not include stop value), step every two values
xticks = [x for x in range(1900, 1920,2)]
print(xticks)


```
