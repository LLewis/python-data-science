```python
import pandas as pd
import numpy as np
import seaborn as sns

print('Hello World')

polls_url = 'https://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'
#reads csv file into a Panda DataFrame
polls = pd.read_csv(polls_url)

print(polls)

#reads csv file into a Panada DataFrame object polls
#object is used to invoke the sort_values and chain the head() method.
#head() method display 5 rows of data.
#polls.sort_values('startdate').head()
polls.sort_values('startdate', ascending=False, inplace=True)

polls.query('state != "U.S."')\
    .plot(x='startdate', y=['adjpoll_clinton', 'adjpoll_trump'])

polls.query('state != "U.S."')\
    .plot(x='startdate', y=['cycle', 'poll_id'])
polls.query('state != "U.S."')\
    .plot(x='startdate', y=['createddate','question_id'])


```
